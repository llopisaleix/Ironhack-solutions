# Data Cleaning 

#### 1. Import pandas library.

```python
import pandas as pd
```

#### 2. Import pymysql and sqlalchemy as you have learnt in the lesson of importing/exporting data. 


```python
import pymysql
from sqlalchemy import create_engine
```

#### 3. Create a mysql engine to set the connection to the server. Check the connection details in [this link](https://relational.fit.cvut.cz/dataset/Stats).

```python
db_engine = create_engine('mysql+pymysql://guest:relational@relational.fit.cvut.cz/stats')
```

#### 4. Import the users table.

```python
users = pd.read_sql_query('SELECT * FROM users;', db_engine)
users.head()
```

#### 5. Rename Id column to userId.

```python
users = users.rename(columns={'Id':'userId'})
users.head()
```

#### 6. Import the posts table. 

```python
posts = pd.read_sql_query('SELECT * FROM posts;', db_engine)
posts.head()
```

#### 7. Rename Id column to postId and OwnerUserId to userId.

```python
posts = posts.rename(columns={'Id':'postId','OwnerUserId':'userId'})
posts.head()
```

#### 8. Define new dataframes for users and posts with the following selected columns:
**users columns**: userId, Reputation, Views, UpVotes, DownVotes  
**posts columns**: postId, Score, userID, ViewCount, CommentCount

```python
users_short = users[['userId','Reputation','Views','UpVotes','DownVotes']]
posts_short = posts[['postId','Score','userId','ViewCount','CommentCount']]
posts_short
```

#### 9. Merge the new dataframes you have created, of users and posts. 
You will need to make an inner [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
merged = pd.merge(posts_short, users_short, on='userId')
merged
```

#### 10. How many missing values do you have in your merged dataframe? On which columns?

```python
missing_values = merged.isnull().sum()
missing_values = missing_values[missing_values > 0]
print("We have", missing_values.values[0], "missing values on", missing_values.index[0], "column.")
merged
```

#### 11. You will need to make something with missing values.  Will you clean or filling them? Explain. 
**Remember** to check the results of your code before going to the next step.

```python
print("If the CommentCount and the Score is zero, or if the Score is negative I would changed it to Zero (0). \nThis probably would mean that there was no intereaction whatsoever with the post. Either no audience or banned post")

merged.loc[(merged['CommentCount']==0) & (merged['Score']==0) & merged['ViewCount'].isnull(), 'ViewCount'] = 0
merged.loc[(merged['Score'] < 0) & merged['ViewCount'].isnull(), 'ViewCount'] = 0

print("\nIf one of the record have values defined I would clean them. Something went wrong while grabing the data and it would be difficult to 'predict' the actual ViewCount.")

to_drop = merged.loc[((merged['CommentCount'] > 0) | (merged['Score'] > 0)) & (merged['ViewCount'].isnull())].index
merged = merged.drop(to_drop,axis=0)

merged
missing_values = merged.isnull().sum()
missing_values
```
