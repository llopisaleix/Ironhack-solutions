```python
# your code here
html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")

#tbody is the whole body of earthquakes and each tr refers to every earthquake
earthquakes = soup.find('tbody', {'id': 'tbody'}).find_all("tr")
#Why do we put this in a variable and not into the for loop directly? Imagine that you want to change this for a report. 
number_of_results = 20
latest_earthquakes = []

#Let's analyze the first object
#print(earthquakes[0])

#In this code we'll use multiple assignation. Doubts about it? -> https://bit.ly/2Xxwi9x

for earthquake in earthquakes[:number_of_results]:
    
    #Date and time are located in between the 'a' tag into the 'tabev6' class
    #split method extracted from Stackoverflow https://bit.ly/3d9MQLo
    
    date,time = earthquake.find('td', {'class': 'tabev6'}).find('a').text.split()
    
#     # Latitude and longitude
#     # First I get the lat and longitude degrees (they are in two different but with the same name tags)
    lat_deg, lon_deg = earthquake.find_all('td', {'class': 'tabev1'})
    lat_dir, lon_dir, magnitude = earthquake.find_all('td', {'class': 'tabev2'})
    lat_deg = f"{lat_deg.text.strip()} {lat_dir.text.strip()}"
    lon_deg = f"{lon_deg.text.strip()} {lon_dir.text.strip()}"
    
    # Region
    region = earthquake.find('td', {'class': 'tb_region'}).text.strip()
#     # Create list of information and append to my original list
    earthquake_summary = [date, time, lat_deg , lon_deg, region]
    latest_earthquakes.append(earthquake_summary)
   
df = pd.DataFrame(latest_earthquakes, columns=['Date', 'Time', 'Latitude', 'Longitude', 'Region'])
df


```
