#### Select all the customers from Spain.
**Hint**: Remember that you are not asked to find orders from Spain but customers. A customer might have more than one order associated. 

```python
# your code here
#Creating a variable setting a column condition
spain_orders = orders[orders.Country == 'Spain']
spain_orders
#second variable getting all the values from customer id and getting the unique values from it:
spain_customer_ids = spain_orders['CustomerID'].unique()
spain_customer_ids
```

#### How many customers do we have in Spain?

```python
# your code here
print(len(spain_customer_ids))
```

#### Select all the customers who have bought more than 50 items.
**Hint**: Remember that you are not asked to find orders with more than 50 items but customers who bought more than 50 items. A customer with two orders of 30 items each should appear in the selection.

```python
# your code here

#Variable to group the customers, in this case the summ applies to all the columns although we're only using quantity after
customers_group = orders.groupby('CustomerID').agg('sum')

# #Now we select from the prevoius variable those customers with > 50 items
big_customers = customers_group[customers_group['Quantity'] > 50].reset_index()
big_customers[['CustomerID','Quantity']].head()
```

#### Select orders from Spain that include more than 50 items.

```python
# your code here
#In spain_customer_ids we have the id's from the customers in spain and in big customers all the customers with > 50 items
big_spain_orders = big_customers[big_customers['CustomerID'].isin(spain_customer_ids)]
big_spain_orders.head()
```

#### Select all free orders.

```python
# your code here
#This one has a bit of exploratory analysis. In this case we determined the value on 0 has to be in amount spent.
free_orders = orders[orders['amount_spent'] == 0]
free_orders
```

#### Select all orders whose description starts with `lunch bag`.
**Hint**: use string functions.

```python
# your code here
# startswith syntax and documentation : https://bit.ly/3etruIR
lunch_bags = orders[orders['Description'].str.startswith("lunch bag")]
lunch_bags.head()
```

#### Select all `lunch bag` orders made in 2011.

```python
# your code here
#There's no known product code for lunch bags. Therefore it has to be done through the description. We will use a very similar 
# to the previous one but changing startswith for contains
lunch_bags = orders[orders['Description'].str.contains("lunch bag")]
#And then we select all those made in 2011
lunch_bags_2011 = lunch_bags[lunch_bags['year'] == 2011]
lunch_bags_2011.head()
```
